```python
from PIL import Image
import numpy as np

image = Image.open(r'Adam_Sandler_1065_542.jpeg').convert('L')
data = np.asarray(image)
data.shape
```

```python
U,S,V = np.linalg.svd(data,full_matrices=False)
```

```python
S= np.diag(S)
```

```python
import matplotlib.pyplot as plt
j=0
for r in (5,20,100,120):
    im=U[:,:r] @ S[0:r,:r] @ V[:r,:]
    plt.figure(j+1)
    j+=1
    imd=plt.imshow(im)
    imd.set_cmap('gray')
    plt.title('r= '+str(r))
    plt.show
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('SV')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))

plt.title('Cumsum')
plt.show()
```
